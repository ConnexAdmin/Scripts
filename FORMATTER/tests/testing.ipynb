```python
import pandas as pd 
import numpy as np
```

```python
df1 = pd.read_csv('Testing_All-Answers---Valley_2023-12-17_00-14-44.csv')
df2 = pd.read_csv('Testing_Valley-Applications_2023-12-17_00-06-35.csv')
```

```python
df1.head()
```

```python
df1 = pd.pivot_table(df1, values='answer-text', index='Name', columns='Question', aggfunc='first')
```

```python
df2["resume"] = df2["resume"].str.slice(start=2)
df2["picture"] = df2["picture"].str.slice(start=2)
```

```python
df3 = df2.merge(df1, on="Name").set_index("Name")
```
